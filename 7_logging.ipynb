### Logging
Logging in python allow us to record and store messages at diffrent severity levels during the execution of our program. It's a powerful tool for understanding the flow of our code, diagonising issues, and monitoring the behaviour of our application.

```python
import logging # to import python's in-built module.
```

#### Configure the logging settings : We can configure the logging module according to our requirements. This includes specifying the logging level, defining the log format, and setting up log handlers

```python
logging.basicConfig(level=logging.DEBUG) # filename='save_log.log',level=logging.DEBUG, filemode='w',format='%(asctime)s - %(levelname)s- %(message)s'
```

1. filename : To define filename where all log infi will be saved.if not defined it will show log messages on notebook.
2. level : To define the minimum order of level should be considered during ececution.
3. filemode : to define file mode it can be only write mode "w" where file will be always save current data and remove all old log info of older excecution and also we can use append "a" where all new info will be appended to log file with removing previous excecutions informations.
4. format : to define the format in which we want to save or show the information in log file.

3. Log massages at diffrent levels : We can use the logging module to log messeges at diffrent levels such as DEBUG, INFO, WARNING, ERROR, or CRITICAL. Each level corresponds to a diffrent severity level of the messages.

```python
logging.debug('This is a debug message.')
logging.info('This is an info message.')
logging.warning('This is a warning message. ')
logging.error('This is an error message.')
logging.critical('This is a critical messages.')
```
